## **Exploratory Data Analysis**
### Descriptive Analysis
Exploring the attributes of the customers, how they interact with each other, and how they interact with the eligibility Status of the Customers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Analysis/data_cleaned.csv')
data.head()
```

```python
data['Gender']=data['Gender'].map({1:'Male',0:'Female'}, na_action=None)
data['Education']=data['Education'].map({1:'Graduate',0:'Not Graduate'}, na_action=None)
data['Married']=data['Married'].map({1:'Yes',0:'No'}, na_action=None)
data['Self_Employed']=data['Self_Employed'].map({1:'Yes',0:'No'}, na_action=None)
data['Property_Area']=data['Property_Area'].map({0:'Rural',1:'Semiurban',2:'Urban'}, na_action=None)
data['Loan_Status']=data['Loan_Status'].map({1:'Y',0:'N'}, na_action=None)
data['Dependents']=data['Dependents'].map({0:'0',1:'1',2:'2',3:'3+'}, na_action=None)
```

**Bivariate Analysis**\
Exploring Relationship between attributes and Loan Status

```python
sns.boxplot(x='Loan_Status',y='LoanAmount',data=data).set_title('Relationship Between Loan Amount and Loan Status')
plt.show()
```

```python
sns.boxplot(x='Loan_Status',y='ApplicantIncome',data=data).set_title("Relationship Between Applicant's Income and Loan Status")
plt.show()
```

```python
sns.boxplot(x='Loan_Status',y='CoapplicantIncome',data=data).set_title("Relationship Between Coapplicant's Income and Loan Status")
plt.show()
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Married'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title('Relationship Marriage Status and Loan Status')
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Gender'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relationship Between Gender and Loan Status")
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Dependents'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relationship Between No. of Dependents and Loan Status")
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Self_Employed'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relationship Between Employment Status and Loan Status")
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Education'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relationship Between Education Status and Loan Status")
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Credit_History'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relatiionship Between Credit History and Loan Status")
```

```python
ct=pd.crosstab(data['Loan_Status'],data['Property_Area'])
ax = ct.plot(kind='bar', stacked=True,figsize=(4,4)).set_title("Relationship Between Property Area and Loan Status")
```
